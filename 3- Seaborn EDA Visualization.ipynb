```python
#SEABORN (EDA İÇİN İLERİ GÖRSELLEŞTİRME)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

```

```python
df = pd.read_csv("titanic.csv")
```

```python
#Barplot (Oran/ortalama karşılaştırma)

plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="2urvived", data=df)
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#Countplot (Frekans sayma)

#Sınıf, kategori dağılımı

plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", data=df)
plt.title("Passenger Count by Class")
plt.show()
```

```python
#Boxplot (Dağılım + outlier)

plt.figure(figsize=(7,4))
sns.boxplot(x="Pclass", y="Age", data=df)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
#Histplot (Dağılım)

plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
df = df.loc[:, ~df.columns.str.contains("zero")]
# gereksiz kolonları silelim
```

```python
#Heatmap (Correlation)

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
