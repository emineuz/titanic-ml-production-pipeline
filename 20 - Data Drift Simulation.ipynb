###  Data Drift Simulation

In this notebook, we simulate distribution shifts in the test data to evaluate how sensitive the model is to potential data drift scenarios.

```python
import joblib
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score, f1_score
```

```python
model = joblib.load("calibrated_rf.pkl")
threshold = joblib.load("best_cost_threshold.pkl")

X_test = joblib.load("X_test.pkl")
y_test = joblib.load("y_test.pkl")
```

```python
y_proba_original = model.predict_proba(X_test)[:,1]
y_pred_original = (y_proba_original >= threshold).astype(int)

acc_original = accuracy_score(y_test, y_pred_original)
f1_original = f1_score(y_test, y_pred_original)

print("Original Accuracy:", acc_original)
print("Original F1:", f1_original)
```

```python
X_drift = X_test.copy()

X_drift["Age"] = X_drift["Age"] + 10
```

```python
y_proba_drift = model.predict_proba(X_drift)[:,1]
y_pred_drift = (y_proba_drift >= threshold).astype(int)

acc_drift = accuracy_score(y_test, y_pred_drift)
f1_drift = f1_score(y_test, y_pred_drift)

print("Drift Accuracy:", acc_drift)
print("Drift F1:", f1_drift)
```

```python
print("Accuracy Change:", acc_original - acc_drift)
print("F1 Change:", f1_original - f1_drift)
```

### To simulate data drift, passenger ages were artificially increased by 10 years. Interestingly, the model's performance slightly improved under this shift, with both accuracy and F1 score increasing marginally. This suggests that the model is relatively robust to moderate distribution changes and that small shifts in the age distribution do not negatively impact predictive performance. The overall stability of the metrics indicates that the model does not rely excessively on a narrow age range.
