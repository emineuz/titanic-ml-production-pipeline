```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
```

```python


def age_group(age):
    if age < 12:
        return "Child"
    elif age < 18:
        return "Teen"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["AgeGroup"] = df["Age"].apply(age_group)
```

```python
df = pd.get_dummies(df, columns=["AgeGroup"], drop_first=True)
```

```python
df["FamilySize"] = df["sibsp"] + df["Parch"] + 1
```

```python
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
```

### yolcunun yalnız mı yoksa ailesi mi var şeklinde yeni feature'lar oluşturuldu


```python
import numpy as np
df["Fare_log"] = np.log1p(df["Fare"])
```
