```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#TREND GÖRMEK (LİNE PLOT)

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 18]

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Basit Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
df = pd.read_csv("titanic.csv")
```

```python
#KARŞILAŞTIRMA (BAR PLOT)

survival_by_sex = df.groupby("Sex")["2urvived"].mean()

plt.figure(figsize=(6,4))
plt.bar(survival_by_sex.index, survival_by_sex.values)
plt.title("Cinsiyete Göre Hayatta Kalma Oranı")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# İLİŞKİYİ (KORELASYON) GÖRMEK (SCATTER PLOT)

plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

```

```python
plt.figure(figsize=(7,4))
plt.bar(survival_by_sex.index, survival_by_sex.values, color=["pink","steelblue"])
plt.title("Survival Rate by Sex")
plt.ylabel("Rate")
plt.xlabel("Sex")
plt.ylim(0,1)
plt.show()
```

```python
df.groupby("Sex")["2urvived"].mean().plot(kind="bar", figsize=(6,4), title="Survival by Sex")

plt.ylabel("Rate")
plt.xlabel("Sex")
plt.show()
```

```python
df.groupby("Pclass")["2urvived"].mean().plot(kind="bar", figsize=(6,4), title="Survival by Class")

plt.ylabel("Rate")
plt.xlabel("Pclass")
plt.show()
```

```python
# Dağılım(Histogram) ==> bins == Histogram aralık sayısı

plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=40)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
